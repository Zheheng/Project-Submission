```python
from bs4 import BeautifulSoup
import urllib
from selenium import webdriver
from time import sleep
from scrapy.selector import Selector
from scrapy.http import HtmlResponse
from selenium.webdriver.chrome.options import Options
import numpy as np
import pandas as pd
import math


chrome_options = Options()
chrome_options.add_argument("--headless")
chromedriver = "/Windows/chromedriver.exe"
```

```python
search_terms = ['data analyst','data engineer','data scientist','business intelligence','data architect',
                'machine learning','deep learning','Natural Language Processing','business analytics',
                'data science','BI','NLP','solution architecture','data mining','predictive analytics','database'
                ,'analytics']

search_list = []
for term in search_terms:
    url = "https://www.mycareersfuture.sg/search?search={}&sortBy=new_posting_date&page={}".format(term,"{}")
    search_list.append(url)
    
search_list
```

```python
url_list =[]
for url in search_list:
    driver = webdriver.Chrome(chromedriver,chrome_options=chrome_options)
    driver.get(url.format(0))
    sleep(3)
    html = driver.page_source
    driver.close
    
    str = Selector(text=html).xpath('//span[@id="search-results-message"]/div/text()').extract()
    post_num = [int(s) for s in str[0].split() if s.isdigit()]
    
    num_pages = math.ceil(int(post_num[0])/20)
    
    for page in range(0,num_pages):
        driver = webdriver.Chrome(chromedriver,chrome_options=chrome_options)
        driver.get(url.format(page))
        sleep(3)
        html = driver.page_source
        driver.close
        href_list = Selector(text=html).xpath('//div[@class="card-list"]/div/div/a/@href').extract()
        
        for i in range(len(href_list)):
            url_list.append("https://www.mycareersfuture.sg" + href_list[i])
     
len(url_list)
```

```python
url_list[0]
```

```python
company = []
job_title = []
sal_lower = []
sal_upper = [] 
sal_type = []
skills = []
roles = []
requirements = []
emp_type = []
seniority = []
industry = []
experience = []
for page in url_list:
    print(page)
    driver.get(page)
    sleep(10)
    html = driver.page_source
    driver.close
        
    company_name = Selector(text=html).xpath('//p[@name="company"]/text()').extract()
    title = Selector(text=html).xpath('//h1[@id="job_title"]/text()').extract()
    sal1 = Selector(text=html).xpath('//span[@class="dib"][1]/text()').extract()
    sal2 = Selector(text=html).xpath('//span[@class="dib"][2]/text()').extract()
    sal_t = Selector(text=html).xpath('//span[@class="salary_type dib f5 fw4 black-60 pr1 i pb"]/text()').extract()
    #r_r = Selector(text=html).xpath('//div[@id="description-content"]//text()').extract()
    #req = Selector(text=html).xpath('//div[@id="requirements-content"]//text()').extract()
    req = Selector(text=html).xpath('//div[@class="jobDescription w-100 v-top relative"]//text()').extract()
    req_skill =Selector(text=html).xpath('//div[@class="multi-pill-button"]/div[@class = "pill-button dib ma1"]/label/text()').extract()
    e_type = Selector(text=html).xpath('//p[@id="employment_type"]/text()').extract()
    level = Selector(text=html).xpath('//p[@id="seniority"]/text()').extract()
    ind = Selector(text=html).xpath('//p[@id="job-categories"]/text()').extract()
    min_exp = Selector(text=html).xpath('//p[@id="min_experience"]/text()').extract()
    
    company.append(company_name)
    job_title.append(title)
    sal_lower.append(sal1)
    sal_upper.append(sal2)
    sal_type.append(sal_t)
    #roles.append(r_r)
    requirements.append(req)
    skills.append(req_skill)
    emp_type.append(e_type)
    seniority.append(level)
    industry.append(ind)
    experience.append(min_exp)
```

```python
data = {'URL':url_list,'company':company,'job_title':job_title,'sal_lower':sal_lower,
        'sal_upper':sal_upper,'sal_type':sal_type,'job_requirements':requirements,
       'skills':skills, 'emp_type':emp_type,'seniority':seniority,
       'industry':industry,'experience':experience}
df = pd.DataFrame.from_dict(data)

#
```
